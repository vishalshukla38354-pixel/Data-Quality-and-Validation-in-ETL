**Question 1 :** Define Data Quality in the context of ETL pipelines. Why is it more than just data cleaning?

**Answer:**

Data Quality in ETL refers to ensuring that data is **accurate, complete, consistent, valid, unique, and timely** before it is loaded into the target system.

It is more than data cleaning because it also includes:

- Data validation against business rules.
- Removing duplicates.
- Maintaining referential integrity.
- Standardizing formats.
- Monitoring data quality continuously.

Data cleaning is only one part of the overall data quality process.

**Question 2 :** Explain why poor data quality leads to misleading dashboards and incorrect decisions.

**Answer:**

Poor-quality data results in inaccurate reports and dashboards, leading to wrong business decisions.

**For example:**

- Duplicate records inflate sales.
- Missing values reduce reporting accuracy.
- Incorrect customer information affects marketing campaigns.
- Invalid dates create incorrect trend analysis.

Therefore, organizations must ensure high-quality data before reporting.

**Question 3 :** What is duplicate data? Explain three causes in ETL pipelines.

**Answer:**

Duplicate data means the same record appears multiple times in a dataset.

**Three causes are:**

1. **Multiple data loads –** The ETL process loads the same file more than once.
2. **Data from multiple sources –** Different systems contain the same customer or transaction.
3. **Missing primary/business key checks –** Records are inserted without checking if they already exist.

**Question 4 :** Differentiate between exact, partial, and fuzzy duplicates.

| Type              | Description                              | Example                              |
| ----------------- | ---------------------------------------- | ------------------------------------ |
| Exact Duplicate   | Every value is identical                 | Rahul Mehta, C101, P11 appears twice |
| Partial Duplicate | Some fields match while others differ    | Same Customer_ID but different city  |
| Fuzzy Duplicate   | Similar values with spelling differences | "Rahul Mehta" and "Rahul Mheta"      |


**Question 5 :** Why should data validation be performed during transformation rather than after loading?

**Answer:**

Validation during transformation:

- Prevents bad data from entering the warehouse.
- Saves storage space.
- Reduces correction effort later.
- Improves ETL performance.
- Ensures reports are generated from valid data.

"Validating after loading may require deleting or correcting already loaded records."

**Question 7 :** Write an SQL query on `Sales_Transactions`  to list all duplicate keys and their counts using the
business key (Customer_ID + Product_ID + Txn_Date + Txn_Amount )

**Answer:**

```python
%%sql
SELECT
    Customer_ID,
    Product_ID,
    Txn_Date,
    Txn_Amount,
    COUNT(*) AS Duplicate_Count
FROM Sales_Transactions
GROUP BY
    Customer_ID,
    Product_ID,
    Txn_Date,
    Txn_Amount
HAVING COUNT(*) > 1;
```

**Output(based on given data):**

| Customer_ID | Product_ID | Txn_Date   | Txn_Amount | Duplicate_Count |
| ----------- | ---------- | ---------- | ---------- | --------------- |
| C101        | P11        | 2025-12-01 | 4000       | 3               |


(The duplicate records are Txn_ID **201, 203, and 208**)

**Question 8 :** Enforcing Referential Integrity

Assume the following `Customers_Master` table:









Identify Sales_Transactions.Customer_ID values that violate referential integrity when joined with
Customers_Master and write a query to detect such violations.


Sales_Transactions
Sales_Transactions
Customers_Mas


**Answer:**

The `Customers_Master` table contains:

- C101
- C102
- C103
- C104

But the `Sales_Transactions` table also contains:

- C105
- C106

These do not exist in `Customers_Master`, so they violate referential integrity.

**SQL Query:**

```python
%%sql
SELECT
    s.Customer_ID
FROM Sales_Transactions s
LEFT JOIN Customers_Master c
ON s.Customer_ID = c.CustomerID
WHERE c.CustomerID IS NULL;
```

**Output:**

| Customer_ID |
| ----------- |
| C105        |
| C106        |


These are the invalid customer IDs because they are not present in the `Customers_Master` table.
